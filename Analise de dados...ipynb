### 1. Importando as bibliotecas

```python
import pandas as pd 
import seaborn as sns
import matplotlib as plt
import datetime
```

### 2. Carregar dados

```python
qualified_leads_raw=pd.read_csv(r'C:\Users\Usuario\Documents\Curso CDO_insta_Fer Face\modulo 04_novo\olist_marketing_qualified_leads_dataset.csv') 
closed_deals_raw=pd.read_csv(r'C:\Users\Usuario\Documents\Curso CDO_insta_Fer Face\modulo 04_novo\olist_closed_deals_dataset.csv') 
```

```python
qualified_leads_raw.head()
```

```python
closed_deals_raw.head()
```

### 3. Carregar cópia e renomear colunas

```python
qualified_leads=qualified_leads_raw.copy()
closed_deals=closed_deals_raw.copy()
```

```python
qualified_leads.shape
```

```python
closed_deals.shape
```

```python
### Renomeando colunas
qualified_leads=qualified_leads.rename(columns={'mql_id':'Marketing Lead id','origin':'origem'})
closed_deals=closed_deals.rename(columns={'mql_id':'Marketing Lead id','sr_id':'sales representative','sdr_id':'sales development person'})
```

```python
closed_deals.head()
```

### 4. Gerar tabelão

```python
tabelao=pd.merge(qualified_leads,closed_deals,on='Marketing Lead id', how='left')
```

```python
tabelao.shape
```

```python
tabelao.head()
```

```python
tabelao.to_excel('tabelao_funil_marketing.xlsx')
```

```python
### colocar as datas em mês (first_contact_date e won_date)
tabelao['month_year_first_contact']=pd.to_datetime(tabelao['first_contact_date']).dt.to_period('M')
tabelao['month_year_won_date']=pd.to_datetime(tabelao['won_date']).dt.to_period('M')
```
